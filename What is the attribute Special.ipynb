## Discovering what is the atributte 'Special'
#### The hypotesys is that special is the sum of some 0-99 atributes

```python
## The set of features we think is related to:
list_Special_Attributes = ['Crossing', 'Finishing', 'HeadingAccuracy','ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
```

```python
specials = players['Special']
```

```python
sum_atts = players[list_Special_Attributes].sum(axis=1)
```

```python
aux_dataframe_specials = pd.DataFrame(specials)
```

```python
aux_dataframe_specials['sum_attributes'] = sum_atts
```

```python
aux_dataframe_specials['special - sum'] = aux_dataframe_specials['Special']-aux_dataframe_specials['sum_attributes']
```

```python
aux_dataframe_specials.describe()
```

#### The result of the cell above indicates that we are adding some feature that is not relevant to the 'Special' attribute

```python
special_investigation = pd.concat([aux_dataframe_specials, players[list_Special_Attributes]], axis=1)
```

```python
special_investigation
```

```python
special_investigation.corr()['special - sum']
```

#### We found that the correlation between 'Composure'  and 'special - sum' = -1
##### It means that Composure shoudn't be summed with other attributes

```python
list_Special_Attributes.remove('Composure')
```

```python
specials = players['Special']
sum_atts = players[list_Special_Attributes].sum(axis=1)
aux_dataframe_specials = pd.DataFrame(specials)
aux_dataframe_specials['sum_attributes'] = sum_atts
aux_dataframe_specials['special - sum'] = aux_dataframe_specials['Special']-aux_dataframe_specials['sum_attributes']
aux_dataframe_specials
```

```python
aux_dataframe_specials.describe()#The result is thar excluding Composure of the sum, we achieve a diference of 0
```
